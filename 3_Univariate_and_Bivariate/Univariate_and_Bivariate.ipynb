## Univariate and Bivariate Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("../cleaned_data.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
data[num_cols].describe().T
```

```python
data[num_cols].hist(figsize=(14, 12), bins=20, edgecolor='black')
plt.suptitle("Univariate Analysis: Numerical Features", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(num_cols[:6]):
    plt.subplot(2, 3, i+1)
    sns.boxplot(data[col], color='skyblue')
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
cat_cols = ['Gender', 'Workout_Type']

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=data, palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

corr['Calories_Burned'].sort_values(ascending=False)
```

```python
numeric_features = ['Age', 'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM',
                    'Resting_BPM', 'Session_Duration (hours)', 'Fat_Percentage',
                    'Water_Intake (liters)', 'Workout_Frequency (days/week)',
                    'Experience_Level', 'BMI']

for col in numeric_features:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=data[col], y=data['Calories_Burned'], color='tomato')
    plt.title(f'{col} vs Calories_Burned')
    plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=col, y='Calories_Burned', data=data, palette='Set3')
    plt.title(f'Calories_Burned vs {col}')
    plt.show()
```

### Key Observations
- `Session_Duration (hours)` and `Avg_BPM` have strong positive correlations with `Calories_Burned`.
- `Workout_Type` shows clear differences  Cardio tends to burn more calories on average.
- `Gender` doesnt show significant variation in calories burned after adjusting for session duration.
- Outliers detected in `Calories_Burned`, `Weight`, and `BMI`, likely from intense workout sessions.
